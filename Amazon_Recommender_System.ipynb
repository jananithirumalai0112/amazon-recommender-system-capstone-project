```python
def get_best_n_sellers(df, category, n):
    """
    Get the top n sellers for a given category.
    :param df: The cleaned DataFrame.
    :param category: The category to filter by (string).
    :param n: The number of top sellers to return.
    :return: A DataFrame of the top n sellers in the specified category.
    """
    top_sellers = df.filter(df["Category"] == category) \
                    .orderBy("SalesRank") \
                    .limit(n)
    return top_sellers

# Example usage
category = "Books"  # You can change the category here
n = 10  # Top 10 sellers
top_books = get_best_n_sellers(df_final, category, n)

# Show the top 10 sellers
top_books.show(10)
```

```python
def get_best_n_sellers(df, category, n=10):
    # Filter data by category
    category_df = df[df['Category'] == category]

    # Sort by AvgRating (highest first) and get top n
    top_sellers = category_df.nlargest(n, 'AvgRating')

    return top_sellers[['ProductID', 'Title', 'AvgRating']]
```

```python
import pandas as pd

data = {
    'UserID': ['User1', 'User2', 'User1', 'User3', 'User2'],
    'ProductID': ['0771044445', '0771044445', '0827229534', '0771044445', '0486220125']
}

df = pd.DataFrame(data)

co_purchase_count = df.groupby('ProductID')['UserID'].nunique()

print(co_purchase_count)
```

```python
# For each product, find the users who co-purchased it
for product in co_purchase_count.index:
    print(f"Product {product} was co-purchased by the following users:")
    users_for_product = df[df['ProductID'] == product]['UserID'].unique()
    print(users_for_product)
    print()
```
